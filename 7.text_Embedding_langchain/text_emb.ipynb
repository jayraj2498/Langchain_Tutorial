Example 1: Semantic Search

Semantic search involves searching for documents or pieces of text based on their meaning rather than exact keyword matches.

```python
# pip install openai langchain numpy  

import openai
from langchain.embeddings import OpenAIEmbeddings
from langchain_openai import OpenAIEmbeddings  
import numpy as np
```

```python
documents = [
   "LangChain is a framework for developing applications powered by language models.",
   "Machine learning enables computers to learn from data.",
   "Natural Language Processing (NLP) involves analyzing and understanding human language.",
]

# Generate embeddings for the documents
doc_embeddings = [embeddings.embed_query(doc) for doc in documents]

# Query text
query = "What is LangChain?"

# Generate embedding for the query
query_embedding = embeddings.embed_query(query)

# Compute similarity (e.g., cosine similarity)
import numpy as np

def cosine_similarity(a, b):
   return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

similarities = [cosine_similarity(query_embedding, doc_emb) for doc_emb in doc_embeddings]

# Find the most similar document
most_similar_index = np.argmax(similarities)
print(f"Most similar document: {documents[most_similar_index]}")
```

Example 2: Text Clustering

Clustering involves grouping similar pieces of text together.

```python
from sklearn.cluster import KMeans

# Sample texts
texts = [
   "LangChain helps with language model applications.",
   "Deep learning is a subset of machine learning.",
   "NLP is used for text processing.",
   "Frameworks like LangChain are useful.",
   "Machine learning models require training data.",
]

# Generate embeddings for the texts
text_embeddings = [embeddings.embed(text) for text in texts]

# Perform KMeans clustering
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(text_embeddings)

# Print the clusters
for i in range(num_clusters):
   cluster_texts = [texts[j] for j in range(len(texts)) if clusters[j] == i]
   print(f"Cluster {i}:")
   for text in cluster_texts:
       print(f" - {text}")
```

Example 3: Document Similarity Comparison

Comparing the similarity between two documents:

```python
doc1 = "LangChain provides utilities to work with language models."
doc2 = "LangChain is a useful tool for language model applications."

# Generate embeddings for the documents
doc1_embedding = embeddings.embed(doc1)
doc2_embedding = embeddings.embed(doc2)

# Calculate cosine similarity
similarity = cosine_similarity(doc1_embedding, doc2_embedding)
print(f"Similarity between doc1 and doc2: {similarity}")
```
