### in document loader , here we can load --> text , pdf , html  , json  filese & more ...

- use documnet loader to load data from source as document 
- document loader provide "Load" method  for loading the data  as documents 

After loading the documents using the appropriate loader, LangChain represents them as a list of Document objects. Each Document object contains the text content and metadata (if available) of the original document. These Document objects can then be processed by other components of LangChain, such as text splitters, embeddings, and language models.

```python
from langchain_openai import ChatOpenAI 
from langchain.prompts import ChatPromptTemplate ,SystemMessagePromptTemplate , HumanMessagePromptTemplate  
```

```python
# in the text loader it can be pdf , txt , csv file  


loader=TextLoader('E:\\Langchain_Prac\\5.document_loader\\sample.txt') 

my_context = loader.load()[0].page_content   

human_template = "{question}\n{company_legal_doc}"  

chat_prompt = ChatPromptTemplate.from_messages([
    HumanMessagePromptTemplate.from_template(human_template) 
])

formated_chat_prompt =chat_prompt.format_messages( 
question = "how can i apply for pancard  " , 
company_legal_doc = my_context   )  


print("formated_chat_prompt is " , formated_chat_prompt) 
```

```python
response= chat.invoke(formated_chat_prompt ) 
print(response)
```

```python
# Word Document Loader:
# This loader can read and load the content of Microsoft Word documents (.docx files).
# It requires the python-docx library to be installed. 


from langchain.document_loaders import Docx2txtLoader

# Load a local Word document
loader = Docx2txtLoader('E:\\Langchain_Prac\\5.document_loader\\SQL QUERIES FOR IPL DATA.docx')
documents = loader.load()
```

```python
# Web Page Loader:
# This loader can load the content of a web page from a given URL.  


from langchain.document_loaders import WebBaseLoader

# Load a web page
loader = WebBaseLoader('https://en.wikipedia.org/wiki/Swiggy')
documents = loader.load()
```

```python
# Print the content of each document
for doc in documents:
    print(doc.page_content)
    print('-' * 80)
```
